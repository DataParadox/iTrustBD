# Degree Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
from scipy import special

fig = plt.figure(figsize = (12, 8))
dist = nodeTable['Degree'].value_counts(normalize=True).sort_index()
plt.title('Degree Distribution in log-log Scale of BTC-Alpha')
plt.ylabel('Probability Distribution, $P_{k}$')
plt.xlabel('Degree, k')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0),useMathText=True)
#plt.plot(dist.index, list(math.log(x) for x in dist.values), 'm.', label = 'BTC-Alpha')
plt.loglog(dist.index, list(x for x in dist.values), 'm.', label = 'BTC-Alpha')

# lambda_ = 0.001
# x_min = 1
gamma = 2.0
C = scipy.special.zeta(gamma)
plt.loglog(dist.index, list(math.pow(x, -gamma)/C for x in dist.index), 'g--', label = '$\gamma = 2.0$')

gamma = 2.5
C = scipy.special.zeta(gamma)
plt.loglog(dist.index, list(math.pow(x, -gamma)/C for x in dist.index), 'b--', label = '$\gamma = 2.5$')

gamma = 3.0
C = scipy.special.zeta(gamma)
plt.loglog(dist.index, list(math.pow(x, -gamma)/C for x in dist.index), 'r--', label = '$\gamma = 3.0$')

# gamma = 2.5
# C = math.pow(lambda_, 1-gamma)/max(1-gamma, lambda_*x_min)
# plt.plot(dist.index, list(C*math.pow(x, -gamma)*math.exp(-lambda_*x) for x in dist.index), 'b--', label = '$\gamma = 2.5$')

# gamma = 2.7
# C = math.pow(lambda_, 1-gamma)/max(1-gamma, lambda_*x_min)
# plt.plot(dist.index, list(C*math.pow(x, -gamma)*math.exp(-lambda_*x) for x in dist.index), 'r--', label = '$\gamma = 2.7$')

plt.legend()
plt.grid(True, linestyle =':')
plt.savefig('degree_distributions.png', transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy
from scipy import special

fig = plt.figure(figsize = (12, 8))
dist = nodeTable['Degree'].value_counts(normalize=True).sort_index()
dist = dist.cumsum()

plt.title('Cumulative Degree Distribution in log-log Scale of BTC-Alpha')
plt.ylabel('Cumulative Distribution, $P_{k}$')
plt.xlabel('Degree, k')
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0),useMathText=True)
plt.loglog(dist.index, list(x for x in dist.values), 'm.', label = 'BTC-Alpha')

gamma = 2.0
C = scipy.special.zeta(gamma)
y = pd.DataFrame(x**(-gamma)/C for x in dist.index)
y = y.cumsum()
plt.loglog(dist.index, y.values, 'g--', label = '$\gamma = 2.0$')

gamma = 2.5
C = scipy.special.zeta(gamma)
y = pd.DataFrame(x**(-gamma)/C for x in dist.index)
y = y.cumsum()
plt.loglog(dist.index, y.values, 'b--', label = '$\gamma = 2.5$')

gamma = 3.0
C = scipy.special.zeta(gamma)
y = pd.DataFrame(x**(-gamma)/C for x in dist.index)
y = y.cumsum()
plt.loglog(dist.index, y.values, 'r--', label = '$\gamma = 3.0$')

plt.legend()
plt.grid(True, linestyle =':')
plt.savefig('cumulative_degree_distributions.png', transparent=True)
plt.show()
```
